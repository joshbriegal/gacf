```python
def get_phase_amplitude(obj, period, epoch=0.0):
    
    timeseries = obj.timeseries_binned
    data = obj.flux_binned
    
    phase_app, data_app = append_to_phase(create_phase(timeseries, period, epoch), data)
    binned_phase_app, binned_data_app = bin_phase_curve(phase_app, data_app)
    phase_amplitude = max(binned_data_app) - min(binned_data_app)
    return phase_amplitude
```

```python
def get_aliases(freq, sampling_freq, lower_limit=None, upper_limit=None, num_aliases=3):
    aliases = []
    if lower_limit is None:
        lower_limit = freq - num_aliases * sampling_freq
    if lower_limit < 0:
        lower_limit = 0
    if upper_limit is None:
        upper_limit = freq + num_aliases * sampling_freq
    calc_plus = True
    calc_minus = True
    i = 1
    aliases.append(freq)
    while calc_plus or calc_minus:
        d = i * sampling_freq
        l = freq - d
        u = freq + d
        if l > lower_limit:
            aliases.append(l)
        else:
            calc_minus = False
        if u < upper_limit:
            aliases.append(u)
        else:
            calc_plus = False
        i += 1
    return aliases
```
